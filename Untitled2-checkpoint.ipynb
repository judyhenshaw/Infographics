```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_excel(r'C:\Users\judyh\Downloads\excel files\Book2.xlsx')
```

```python
df=pd.DataFrame(dataset)
```

```python
df
```

```python
x = np.arange(37)
width = 0.4
fig, ax = plt.subplots(figsize=(40,8))
y1 = plt.bar(x, df['UNDEREMPLOYMENT'], width, alpha=0.5, color='#EE3224')
y2 = plt.bar(x+width, df['UNEMPLOYMENT'], width, alpha=0.5, color='#F78F1E')
plt.xlabel('STATE')
plt.ylabel('% of data')
plt.xticks(x+0.2, ['Abia', 'Adamawa', 'Akwa Ibom', 'Anambra', 'Bauchi', 'Bayelsa', 'Benue', 'Borno', 'Cross River', 'Delta', 'Ebonyi', 'Edo', 'Ekiti', 'Enugu', 'Gombe', 'Imo', 'Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Kogi', 'Kwara', 'Lagos', 'Nasarawa', 'Niger', 'Ogun', 'Ondo', 'Osun', 'Oyo', 'Plateau', 'Rivers', 'Sokoto', 'Taraba', 'Yobe', 'Zamfara', 'FCT'])
ax.set_xticklabels(df['state'])
ax.set_title('LABOR FORCE STATISTICS IN NIGERIA Q3 2018')
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(y1)
autolabel(y2)
plt.xlim(min(x)-width, max(x)+width*4)
plt.ylim([0, max(df['UNDEREMPLOYMENT'] + df['UNEMPLOYMENT'])])
plt.legend(['UNDEREMPLOYMENT', 'UNEMPLOYMENT'], loc='upper right')
plt.grid()
plt.show()
```
